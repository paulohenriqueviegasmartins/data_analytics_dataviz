LIST e DICT COMPREHENSION

```python
lista = [30, 40, 50, 60]
[1 for x in lista] # List comprehension
```

```python
lista = [3000, 4000, 50, 60]
[x for x in lista] # List comprehension
```

```python
lista = [30, 40, 50, 60]
[x*2 for x in lista] # List comprehension
```

```python
lista = [30, 40, 50, 60]
[" " for x in lista] # List comprehension
```

```python
lista = [30, 40, 50, 60]
["Vai dar certo" for x in lista] # List comprehension
```

```python
[x for x in lista if x%2!=0]
```

```python
# Dict Comprehension
nomes = ["Arthur", "Paulo", "Ana"]
{n: len(n) for n in nomes}
```

```python
# Criar um dicionário
{c: df[c].dropna().max() for c in df.columns}
```
